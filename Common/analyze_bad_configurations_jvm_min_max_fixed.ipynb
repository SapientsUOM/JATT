```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
filename = 'cse-B85-HD3.db_startup.scimark.sparse_less_than_40'
```

```python
#read conigurations
scimark_sor = pd.read_csv('bad_configurations/'+filename+'.csv')
```

```python
#get parameter value flags
scimark_sor_param = scimark_sor[scimark_sor.type != 'bool']
scimark_sor_param = scimark_sor_param.drop('type', axis=1)
```

```python
# convert each value to a numeric type

def convert_to_int(string):
    value = 0
    try:
        value = int(string)
    except TypeError:
        value = float(string)
    finally:
        return value
```

```python
columns = scimark_sor_param.columns
columns = columns[1:]
for col in columns:
    scimark_sor_param[col] = scimark_sor_param[col].apply(convert_to_int)
```

```python
scimark_sor_param = scimark_sor_param[scimark_sor_param.default_initial_value != 0]
```

```python
#get min max values for flags
def get_max(default):
    max = default
    if default>=0:
        max = math.ceil(default + (default/2.0))
    else:
        max = math.ceil(default - (default/2.0))
    return max

def get_min(default):
    min = default
    if default>=0:
        min= math.floor(default - (default/2.0))
    else:
        min = math.floor(default + (default/2.0))
    return min


scimark_sor_param['maximum'] = scimark_sor_param.default_initial_value.apply(get_max)
scimark_sor_param['minimum'] = scimark_sor_param.default_initial_value.apply(get_min)
```

```python
scimark_sor_param.drop('default_initial_value', axis=1, inplace=True)
```

```python
flag_normalized_df=pd.DataFrame()
for row in scimark_sor_param.iterrows():
    #print row[1][0]
    size = len(scimark_sor_param.columns)
    minimum = row[1][size-1]
    maximum = row[1][size-2]
    numeric_values = row[1][1:size-2]
    numeric_values_array = np.array(numeric_values)
    
    if(minimum == maximum):
        print 'error'
    
    num_val=[]
    for v in numeric_values_array:
       v=(float(v)-minimum)/math.fabs((maximum-minimum))
       num_val.append(v)
       #print v
    flag_normalized_df[row[1][0]]=num_val
    
```

```python
# Plot it out
fig, ax = plt.subplots()
heatmap = ax.pcolor(flag_normalized_df, cmap=plt.cm.Reds, alpha=0.9)

##################################################
## FORMAT ##
##################################################

fig = plt.gcf()
fig.set_size_inches(50,50)

# turn off the frame
ax.set_frame_on(True)

# put the major ticks at the middle of each cell
ax.set_yticks(np.arange(flag_normalized_df.shape[0])+0.5, minor=False)
ax.set_xticks(np.arange(flag_normalized_df.shape[1])+0.5, minor=False)

# want a more natural, table-like display
ax.invert_yaxis()
ax.xaxis.tick_top()
 
# Set the labels
labels = flag_normalized_df.columns
labels = labels.tolist()
print type(labels)
```

```python


ax.set_xticklabels(labels, minor=False) 
ax.set_yticklabels(flag_normalized_df.index, minor=False)

# rotate the 
plt.xticks(rotation=90)

ax.grid(False)

# Turn off all the ticks
ax = plt.gca()
```
